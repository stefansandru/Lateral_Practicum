# Task

```python
class Task:
  def __init__(self, description: str, done: bool = False):
    self.description = description
    self.done = done

  def __repr__(self) -> str:
    status = "[X] " if self.done else "[ ] "
    return f"{status}{self.description}"
```

# Repo

```python
from typing import List

class Repo:
  def __init__(self) -> None:
    self.todo_list: List[Task] = []

  def _validate_index(self, task_index: int) -> None:
    """
    Internal helper to validate 0-based index.
    """
    if not isinstance(task_index, int):
      raise TypeError("Index must be an integer.")
    if task_index < 0 or task_index >= len(self.todo_list):
      raise IndexError(f"Index out of range. Valid range: 1..{len(self.todo_list)}")

  def add_task(self, task_desc: str) -> None:
    """
    Add a new Task object to the to-do list.
    """
    if task_desc is None or not str(task_desc).strip():
      raise ValueError("Task description cannot be empty.")
    clean = str(task_desc).strip()
    self.todo_list.append(Task(clean))

  def mark_task_complete(self, task_index: int) -> None:
    """
    Mark a task as complete by setting its done attribute to True.
    Expects a 0-based index.
    """
    self._validate_index(task_index)
    task = self.todo_list[task_index]
    if task.done:
      # Already complete;
      return
    task.done = True

  def remove_task(self, task_index: int) -> str:
    """
    Remove a task by its 0-based index and return the removed task's string representation.
    """
    self._validate_index(task_index)
    removed_task = self.todo_list.pop(task_index)
    return repr(removed_task)
```

# CLI

```python
class CLI:
  def __init__(self, repo: Repo) -> None:
    self.repo = repo

  def view_tasks(self) -> str:
    """
    Return a human-readable string of all tasks with 1-based indices.
    Also prints it.
    """
    if not self.repo.todo_list:
      output = "No tasks yet. Add one!\n"
      print(output, end="")
      return output

    lines = ["Your To-Do List:"]
    for idx, task in enumerate(self.repo.todo_list, start=1):
      lines.append(f"{idx}. {task}")
    output = "\n".join(lines) + "\n"
    print(output)
    return output

  def _print_menu(self) -> None:
    print("\nTo-Do List Manager")
    print("-" * 20)
    print("1) Add a task")
    print("2) View tasks")
    print("3) Mark a task complete")
    print("4) Remove a task")
    print("5) Exit")

  def _prompt_int(self, prompt: str) -> int:
    txt = input(prompt).strip()
    return int(txt)

  def main(self) -> None:
    """
    command-line loop
    """
    while True:
      self._print_menu()
      choice = input("Choose an option (1-5): ").strip()
      if choice == "1":
        task_desc = input("Enter task description: ")
        try:
          self.repo.add_task(task_desc)
          print("Task added.")
        except ValueError as e:
          print(f"Error: {e}")
      elif choice == "2":
        self.view_tasks()
      elif choice == "3":
        if not self.repo.todo_list:
          print("Nothing to complete. Your list is empty.")
          continue
        try:
          self.view_tasks()
          idx = self._prompt_int("Enter the task number to mark complete: ")
          self.repo.mark_task_complete(idx - 1)  # convert to 0-based
          print("Task marked as complete.")
        except (ValueError, IndexError, TypeError) as e:
          print(f"Error: {e}")
      elif choice == "4":
        if not self.repo.todo_list:
          print("Nothing to remove. Your list is empty.")
          continue
        try:
          self.view_tasks()
          idx = self._prompt_int("Enter the task number to remove: ")
          removed = self.repo.remove_task(idx - 1)  # convert to 0-based
          print(f"Removed: {removed}")
        except (ValueError, IndexError, TypeError) as e:
          print(f"Error: {e}")
      elif choice == "5":
        print("Goodbye!")
        break
      else:
        print("Invalid option. Please enter a number from 1 to 5.")
```

# Run

```python
repo = Repo()
cli = CLI(repo)
cli.main()
```
